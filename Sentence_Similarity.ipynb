```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer
```

```python
# Load the datasets
tweets = pd.read_csv('twitter_sentences.csv',engine="python")
mp = pd.read_csv('mp_sentences.csv',engine="python")
keywords = pd.read_csv('keywords.csv',engine="python")



```

```python
# Download a pre-trained BERT model (e.g., 'bert-base-nli-mean-tokens')
model = SentenceTransformer('bert-base-nli-mean-tokens')

# Assuming 'x' is the column you want to concatenate in the DataFrame 'mp'
sentences_text = keywords['x'].tolist()

# Concatenate all texts into one big text
big_text = ' '.join(sentences_text)

tweets_text = tweets
#
```

```python
#Encode the text into embeddings using BERT
tweets_embeddings = model.encode(tweets_text, convert_to_tensor=True)
mp_embedding = model.encode(big_text, convert_to_tensor=True)

```

```python
# Initialize an empty list to store the most similar sentence index for each tweet
cosine_similarities = []

# Iterate over each tweet's embedding
for tweet_embedding in tweets_embeddings.cpu():
    # Calculate cosine similarity between the tweet and the single document
    similarity = cosine_similarity(tweet_embedding.reshape(1, -1), sentences_embeddings.cpu().reshape(1, -1))

    # Find the index of the most similar sentence
    cosine_similarities.append(similarity[0][0])


# Create a DataFrame to store tweet text and their corresponding cosine similarities
result_df = pd.DataFrame({'Tweet Text': tweets_text, 'Cosine Similarity': cosine_similarities})

# Sort the DataFrame by cosine similarity in descending order
result_df = result_df.sort_values(by='Cosine Similarity', ascending=False)

# Display the result DataFrame
result_df.head()

```
